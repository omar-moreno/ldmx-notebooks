```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from utils.histograms import histogram1D
```

```python
df = pd.read_csv('single_e_1pt2_gev_1e9_01.csv', index_col=False)
df_4gev = pd.read_csv('single_e_4gev_100k.csv', index_col=False)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['prob'], 0, 1, 101, x_label='Probability', norm=True, label='All');
histogram1D(ax, df_4gev['prob'], 0, 1, 101, x_label='Probability', norm=True, color='red', label='All')
ax.set_yscale('log')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['chi2'], 0, 50, 51, x_label='chi2', label='All');
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['chi2']/df['ndf'], 0, 20, 81, x_label='Chi2/NDF', norm=True, label='1.2 GeV')
histogram1D(ax, df_4gev['chi2']/df_4gev['ndf'], 0, 20, 81, color='red', norm=True, x_label='Chi2/NDF', label='4 GeV');
ax.set_yscale('log')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['d0'], -100, 100, 101, x_label='d0', norm=True, label='1.2 GeV');
histogram1D(ax, df_4gev['d0'], -100, 100, 101, x_label='d0', norm=True, label='4.0 GeV', color='red');
ax.set_yscale('log')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['z0'], -100, 100, 101, x_label='z0', norm=True, label='1.2 GeV');
histogram1D(ax, df_4gev['z0'], -100, 100, 101, x_label='z0', norm=True, label='4.0 GeV', color='red');
ax.set_yscale('log')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['phi'], -.5, .5, 101, x_label='phi', norm=True, label='1.2 GeV');
histogram1D(ax, df_4gev['phi'], -.5, .5, 101, x_label='phi', norm=True, label='4.0 GeV', color='red');
ax.set_yscale('log')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['theta'], 1.2, 1.8, 201, x_label='theta', norm=True, label='1.2 GeV');
histogram1D(ax, df_4gev['theta'], 1.2, 1.8, 201, x_label='theta',  norm=True, label='4.0 GeV', color='red');
ax.set_yscale('log')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10), 
                       facecolor='white')
histogram1D(ax, df['n_hits'], 0, 16, 17, x_label='Hit Multiplicity', norm=True, label='1.2 GeV');
histogram1D(ax, df_4gev['n_hits'], 0, 16, 17, x_label='Hit Multiplicity', norm=True, label='4.0 GeV', color='red');
ax.set_yscale('log')
ax.legend()
```
