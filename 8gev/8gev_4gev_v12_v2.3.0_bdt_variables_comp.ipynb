```python
# BDT variables are stored in CSV files. Load the signal and 
# background datasets into Pandas data frames.
frames = { '8 GeV ECal PN':pd.read_csv('8gev_ecal_pn.csv'), 
           '4 GeV ECal PN':pd.read_csv('4gev_ecal_pn.csv'), 
         }
```

```python
colors = [ "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7" ]

def draw_overlay(frames, variable, x_min, x_max, bins, x_label):
  fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 10), facecolor='white')
  for i, (key, frame) in enumerate(frames.items()):
    pu.draw_histogram1D(ax, frame[variable].to_numpy(), x_min, x_max, bins, norm=1, 
                        edgecolor=colors[i], color=colors[i], errors=True, x_label=x_label, 
                        label=key, type='step')
  
  ax.set_yscale("log")
  ax.legend(loc=0)
    
  return ax
```

```python
ax = draw_overlay(frames, 'nReadoutHits', 0, 300, 301, 'Readout Hit Multiplicity')
ax.set_ylim([1e-7, 3])
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'deepestLayerHit', 0, 35, 36, 'ECal Layer ID')
ax.legend(loc=2);
```

```python
ax = draw_overlay(frames, 'summedDet', 0, 25000, 251, 'Total Energy (MeV)')
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'summedTightIso', 0, 20000, 201, 'Energy in isolation region (MeV)')
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'maxCellDep', 0, 10000, 101, 'Maximum energy deposited in a cell (MeV)')
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'showerRMS', 0, 200, 201, 'Shower RMS')
ax.set_ylim([1e-7, 10])
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'xStd', 0, 200, 201, '$\sigma$(x)')
ax.set_ylim([1e-7, 10])
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'yStd', 0, 200, 201, '$\sigma$(y)')
ax.set_ylim([1e-7, 10])
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'avgLayerHit', 0, 30, 31, 'Average layer hit')
ax.set_ylim([1e-7, 10])
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'stdLayerHit', 0, 20, 21, '$\sigma$(Layer)')
ax.legend(loc=1);
```

```python
ax = draw_overlay(frames, 'ecalBackEnergy', 0, 15000, 151, 'Energy deposited in back of ECal (MeV)')
ax.legend(loc=1);
```

```python
for i in range(0, 5): 
  ax = draw_overlay(frames, 'electronContainmentEnergy_%s' % i, 0, 15000, 151, 
                    'Energy in $e^{-}$ containment region %s (MeV)' % i)
  ax.legend(loc=1);
```

```python
for i in range(0, 5):
  ax = draw_overlay(frames, 'photonContainmentEnergy_%s' % i, 0, 15000, 151, 
                    'Energy in $\gamma$ containment region %s (MeV)' % i)
  ax.legend(loc=1);
```

```python
for i in range(0, 5): 
  ax = draw_overlay(frames, 'outsideContainmentNHits_%s' % i, 0, 250, 251, 
                    'Hit multiplicity outside the containment region %s' % i)
  ax.legend(loc=1);
```
